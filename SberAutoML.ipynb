## 3.2. Формирование датасета визитов с признаком целевого события

```python
# Формируем список полей-признаков целевого события в датасете событий
target_sign = ['sub_car_claim_click', 'sub_car_claim_submit_click',
               'sub_open_dialog_click', 'sub_custom_question_submit_click',
               'sub_call_number_click', 'sub_callback_submit_click',
               'sub_submit_success', 'sub_car_request_submit_click']
```

```python
# Формируем вспомогательное поле для агрегации событий по визитам
df_hits['session'] = df_hits['session_id']
```

```python
# Формируем поле с признаком целевого события, совершённого клиентом
df_hits['conversion_rate'] = df_hits.event_action.\
                                apply(lambda x: 1 if x in target_sign else 0)
```

```python
# Агрегируем датафрейм событий по визитам с указанием наличия целевого события
agg_function = {'conversion_rate': 'max'}
df_hits_ag = df_hits.groupby('session_id').agg(agg_function)
```

```python
# Добавляем в датафрейм визитов признак целевого события
df_sessions = df_sessions.merge(df_hits_ag, how='inner', on='session_id')
```

```python
# Определим список признаков и таргета, по которым должно производиться обучение
# и прогнозирование целевого действия пользователя сайта
fit_target_list = ['utm_source', 'utm_medium', 'utm_campaign', 'utm_adcontent',
            'device_category', 'device_brand', 'device_screen_resolution', 'device_browser',
            'geo_country', 'geo_city', 'conversion_rate']
```

```python
# Отобразим корреляцию признаков и таргета
import seaborn as sns
corr_mat = df_sessions[fit_target_list].apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)
sns.heatmap(corr_mat, cmap='vlag', center=0, square=True, linewidths=.5)
plt.show()
```

**Вывод:** из полученной визуализации корреляции признаков видно что между собой тесно коррелируют признаки источников и рекламных компаний, также с ними есть корреляция отдельных признаков устройств и браудеров. Целевое действие отдельно не коррелирует с каким либо признаком.

## 3.3. Кодирование фичей для загрузки в модель машинного обучения

```python
# Прочитаем из хранилища датасет sessions
connection = bd.connect('session.db')
df = pd.read_sql("SELECT * FROM table_sessions", connection)
connection.close()
```

```python
# Зададим тип категориальных фичей, по которым будет проводиться обучение модели
fit_list = fit_target_list.copy()
fit_list.remove('conversion_rate')
df[fit_list] = df[fit_list].astype('category')
df.info()
```

```python
# Определим количество значений в каждой категориальной фиче
df.select_dtypes('category').nunique().sort_values(ascending=False)
```

```python
# Закодируем категориальные фичи с помощью кодировщика OneHotEncoder
df_encoder = pd.get_dummies(df[fit_list], columns=fit_list, prefix=fit_list)
```

```python
# Получим итоговый датафрейм для обучения (с фичами и таргетом)
df_encoder = df_encoder.join(df['conversion_rate'])
```

```python
# Количество используемых признаков для обучения
df_encoder.nunique().count()
```
