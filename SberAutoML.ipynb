```python
# Расчитываем CR для платного и органического трафиков
df_paid_CR['CR'] = round(100 * df_paid_CR['conversion_rate']\
                         / df_paid_CR['session_id'], 2)
df_organic_CR['CR'] = round(100 * df_organic_CR['conversion_rate']\
                         / df_organic_CR['session_id'], 2)
```

### 6.1.1. Определение источников с наибольшим целевым трафиком

```python
# Агрегируем датафрейм событий по источникам по дням с указанием целевого события
agg_function = {'session_id': 'count', 'conversion_rate': 'sum'}
utm_source_traffic = dict()
for elem in df_sessions.utm_source.unique():
    df_sessions_source = df_sessions.\
                        loc[df_sessions.utm_source == elem].\
                        groupby('visit_date').\
                        agg(agg_function)
    df_sessions_source['CR'] = round((df_sessions_source['conversion_rate'] /\
                               df_sessions_source['session_id']) * 100, 2)
    df_sessions_source.drop(['session_id'], axis=1, inplace=True)
    utm_source_traffic[elem] = df_sessions_source
```

```python
# Приводим датафреймы трафика к единому размеру
# (добавляем нулевые значения трафика в даты когда трафик отсутствовал)
all_index = pd.date_range(min_date, max_date).date
for elem in utm_source_traffic:
    if len(elem) < len(all_index):
        non_data_index = set(all_index) - set(utm_source_traffic[elem].index)
        value = [0.0 for _ in range(len(non_data_index))]
        df_non_index = pd.DataFrame({'date': list(non_data_index), 'conversion_rate': value, 'CR': value})
        df_non_index = df_non_index.set_index(['date'])
        utm_source_traffic[elem] = pd.concat([utm_source_traffic[elem], df_non_index])
```

```python
# Формируем список трафиков для сортировки по наибольшей статистической величине
# (для статистического сравнения оставляем только датафреймы с ненулевым трафиком)
dst = [(name, df) for name, df in utm_source_traffic.items() if df.CR.sum() != 0]
```

```python
# Сортируем спискок трафиков по возрастанию статистической величины CR
for i in range(len(dst) - 1):
    for j in range(len(dst) - i - 1):
        function = choose_method(dst[j][1].CR, \
                         dst[j+1][1].CR,\
                         dependenceFlag=False, check='less')
        if function > 0.05:
            dst[j], dst[j+1] =\
            dst[j+1], dst[j]
```

При этом если клиент за один визит несколько раз посещал страницу одного автомобиля, то повтороное посещение в рамках одного визита не учитывается при расчёте спроса на автомобиль.

Увеличение нашего присутсвия в соцсетях целесообразно если:
1. (**УСЛОВИЕ 1**) Не происходит снижения объёмов целевых действий в визитах из соцсетей или сохранения их объёма на прежнем уровне при росте `визитов из соцсетей`. Сочетание указанных факторов свидетельствуют о том что канал __соцсети__ находится в полном насыщении нашим присутсвием и дальнейшее увеличение участия в нём не улучшит целевой трафик.  
2. Если отсутствуют факторы, указанные в п.1:
 - и (**УСЛОВИЕ 2**) объём `визитов из соцсетей с целевыми действиями` значителен в общем объёме `визитов с целивыми действиями` - тогда увеличение нашего присутсвия в соцсетях - увеличит целевой трафик из соцсетей и значит значительно увеличит общий целевой трафик. Зададимся долей существенностий канала `соцсети` рассчитанной как отношение количества каналов, соответвующих к соцсетям (см. п.2.3. исследования), к общему количеству каналов.

 - и (**УСЛОВИЕ 3**) coversion rate (%) `визитов из соцсетей с целевыми действиями` выше coversion rate (%) `всех визитов с целивыми действиями` - тогда увеличение количества и качества нашего присутсвия в соцсетях - увеличит coversion rate (%) из соцсетей, а занчит значительно увеличит общий coversion rate (%).
>*В исследовании предполагаем что по всем каналам стоимость нашего присутсвия равнозначна, в ином случае при формировании рекомендаций целесообразно учесть стоимость привлечения клиентов в разные каналы*
